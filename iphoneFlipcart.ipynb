```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text,'lxml')
```

```python
print(soup.prettify())
```

## TO FIND OUT THE NAMES OF IPHONE

```python
for i in soup.find_all('div', class_="KzDlHZ"):
    print(i.text.strip())
```

## TO FIND OUT THE RATINGS

```python
len(soup.find_all('div',class_="XQDdHH"))
```

## TO FIND OUT THE NUMBER OF RATINGS AND REVIEWS

```python
spans = soup.find_all('span',class_="Wphh3N")[0]
for span in spans:
    print(span.text)
    s=span.text
    str=s.split('&')
    n_rat = int(str[0].strip().split()[0].replace(',',''))
    n_rev = int(str[1].strip().split()[0].replace(',',''))
    print(n_rat,' ' , n_rev)
```

## CONSIDERING THE WHOLE CONTAINER

```python
phone = soup.find_all('div',class_="tUxRFH")
len(phone)
```

```python
name=[]
rating=[]
n_rating=[]
n_reviews=[]

for i in phone:
    name.append(i.find('div', class_="KzDlHZ").text.strip())
    rating.append(i.find('div',class_="XQDdHH").text.strip())
    for span in i.find('span',class_="Wphh3N"):
        str=span.text.split('&')
        n_rating.append(str[0].strip().split()[0])
        n_reviews.append(str[1].strip().split()[0])

df=pd.DataFrame({'Name': name, 'Rating': rating, 'No of Ratings': n_rating, 'No of Reviews': n_reviews})
```
